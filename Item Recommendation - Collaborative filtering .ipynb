### Finding Similarities

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

##### Finding Similar Users

```python
username =item_user.columns
```

```python
username
```

```python
itemnames=user_item.columns
```

```python
itemnames
```

```python
user_sim = 1 - pairwise_distances(user_item,metric='cosine')
```

```python
user_sim = pd.DataFrame(user_sim,index=username,columns=username)
```

```python
user_sim
```

##### Finding Similar products

```python
item_name = user_item.columns
```

```python
item_name
```

```python
item_sim_df = 1 - pairwise_distances(item_user,metric='cosine')
```

```python
item_sim_df
```

```python
item_sim_df = pd.DataFrame(item_sim_df, columns=item_name,index=item_name)
```

```python
item_sim_df
```

```python
def get_similar_users( userid, topN = 5 ):
    users_df = pd.DataFrame()
    users_df['similarity'] = user_sim[userid]
    top_n = users_df.sort_values( ["similarity"], ascending = False )[1:topN+1]
    print( "Top 5" + " Similar users to: " + userid + " are", )
    return top_n

def get_similar_items( itemid, topN = 5 ):
    items_df = pd.DataFrame()
    items_df['similarity'] = item_sim_df[itemid]
    top_n = items_df.sort_values( ["similarity"], ascending = False )[1:topN+1]
    print( "Top 5 Similar items to: " + itemid + " are", )
    return top_n
```
